```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Part 1
#a
brainFile = './Data/brainsize.csv'
brainFrame = pd.read_csv(brainFile)
#b
brainFrame.head()
```

```python
#Part 2
# Використання методу describe() для огляду даних
brainFrame.describe()
```

```python
# Відокремлення даних за статтю
menDf = brainFrame[(brainFrame.Gender == 'Male')]
womenDf = brainFrame[(brainFrame.Gender == 'Female')]
```

```python
# Побудова діаграми розсіювання для чоловіків
menMeanSmarts = menDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(menMeanSmarts, menDf["MRI_Count"])
plt.title("Зв’язок середнього IQ і розміру мозку (чоловіки)")
plt.xlabel("Середній IQ (PIQ, FSIQ, VIQ)")
plt.ylabel("MRI Count")
plt.show()
```

```python
# Побудова діаграми розсіювання для жінок
womenMeanSmarts = womenDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(womenMeanSmarts, womenDf["MRI_Count"])
plt.title("Зв’язок середнього IQ і розміру мозку (жінки)")
plt.xlabel("Середній IQ (PIQ, FSIQ, VIQ)")
plt.ylabel("MRI Count")
plt.show()
```

```python
#Part 3
# Обчислення кореляції між усіма змінними в наборі даних
brainFrame.corr(numeric_only=True, method='pearson')
```

```python
# Обчислення кореляції лише для жінок
womenDf.corr(numeric_only=True, method='pearson')
```

```python
# Обчислення кореляції лише для чоловіків
menDf.corr(numeric_only=True, method='pearson')

```

```python
# Part 4
# Обчислюємо кореляцію лише для числових стовпців
wcorr = womenDf.select_dtypes(include=np.number).corr()

# Побудова теплової карти кореляцій
sns.heatmap(wcorr, annot=True, cmap='coolwarm')
plt.title("Кореляційна карта (Жінки)")
plt.show()
```

```python
mcorr = menDf.select_dtypes(include=np.number).corr()
sns.heatmap(mcorr, annot=True, cmap='coolwarm')
plt.title("Кореляційна карта (Чоловіки)")
plt.show()
```
